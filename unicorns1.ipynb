```python
# Convert 'Date Joined" column to datetime.

df['Date Joined'] = pd.to_datetime(df['Date Joined'])

# Check datatypes of the dataset

df.dtypes
```

```python
# Create a month joined column

df['Month Joined'] = df['Date Joined'].dt.month_name()
df.head()
```

```python
# Create years to join column (how many years company took to become a unicorn)

df['Years to Join'] = df['Date Joined'].dt.year - df['Year Founded']
df.head()
```

```python
# Filter dataset to the year of 2021

df_2021 = df[df['Date Joined'].dt.year == 2021]

df_2021.head()
```

```python
# Trend overtime observation, week in which companies joined the unicorn list

# Week joined column
df_2021.insert(3, 'Week Joined', df_2021['Date Joined'].dt.strftime('%Y-W%V'), True)

# Group companies by week joined
df_by_week_2021 = df_2021.groupby(by='Week Joined')['Company'].count().reset_index().rename(columns={'Company':'Company count'})

df_by_week_2021.head()
```

```python
#Trend overtime comparison

#filter for one additional year (2020)
df_2020 = df[df['Date Joined'].dt.year == 2020]

#concatenate 2020 and 2021 filtered datasets
df_2020_2021 = pd.concat([df_2020, df_2021.drop(columns='Week Joined')])

#adding a quarter joined column
df_2020_2021['Quarter Joined'] = df_2020_2021['Date Joined'].dt.to_period('Q').dt.strftime('%Y-Q%q')

#converting the 'valuation' column to numeric by removing '$' and 'B' and casting it to float data type
df_2020_2021['Valuation'] = df_2020_2021['Valuation'].str.strip('$B').astype(float)

#group df_2020_2021 by quarter joined and aggregate by average funding
df_by_quarter_2020_2021 = df_2020_2021.groupby(by='Quarter Joined')['Valuation'].mean().reset_index().rename(columns={'Valuation':'Average Valuation'})

df_by_quarter_2020_2021.head()
```

#Visualization

```python
#Time to Unicorn visualization

month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July',
               'August', 'September', 'October', 'November', 'December']
print(month_order)
```

```python
#Distribution of years to become unicorn with respect to month joined

sns.boxplot(x=df['Month Joined'],
            y=df['Years to Join'],
            order=month_order,
            showfliers=False)

#Plot title
plt.title('Distribution of years to become unicorn with respect to month joined')

#Rotate x-axis labels to avoid overlap
plt.xticks(rotation=45, horizontalalignment='right')

plt.show()
```

#Results and Evaluation

```python
#Additional graphic representation, average number of years it took companies to achieve Unicorn status

#size of the plot
plt.figure(figsize=(10,6))

#creating the plot
sns.barplot(x=df['Year Founded'], y=df['Years to Join'], errorbar=('ci', False))

#plot title and labels
plt.title('Bar plot of years to join with respect of year founded')

plt.xlabel('Year Founded')

plt.ylabel('Years to join Unicorn Status')

plt.xticks(rotation=45, horizontalalignment='right')

plt.show
```

```python
#Visualizing number of companies that joined per interval

#Figure size
plt.figure(figsize = (20,5))

#Creating the bar plot
plt.bar(x=df_by_week_2021['Week Joined'], height=df_by_week_2021['Company count'])
plt.plot()

#Title and labels
plt.xlabel('Week Number')

plt.ylabel('Number of Companies')

plt.title('Number of companies that became Unicorns per week in 2021')

plt.xticks(rotation=45, horizontalalignment='right')

plt.show()
```

```python
#Visualize the average valuation over the quarters

df_by_quarter_2020_2021['Quarter Number'] = df_by_quarter_2020_2021['Quarter Joined'].str[-2:]

df_by_quarter_2020_2021['Year Joined'] = df_by_quarter_2020_2021['Quarter Joined'].str[:4]

#Creating the plot
plt.figure(figsize = (10,5))

sns.barplot(x=df_by_quarter_2020_2021['Quarter Number'],
            y=df_by_quarter_2020_2021['Average Valuation'],
            hue=df_by_quarter_2020_2021['Year Joined'])

plt.plot()

#Title and labels
plt.xlabel('Quarter Number')

plt.ylabel('Average Valuation (in USD billions)')

plt.title('Average Valuation of Companies that became Unicorns in 2020 vs 2021')

plt.show()
```
